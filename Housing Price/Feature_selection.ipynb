```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns', None)
```

```python
X_train = pd.read_csv('xtrain.csv')
X_test = pd.read_csv('xtest.csv')
X_train.head()
```

```python
y_train = X_train['SalePrice']
y_test = X_test['SalePrice']
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.drop(['Id','SalePrice'],axis=1,inplace=True)
X_test.drop(['Id','SalePrice'],axis=1,inplace=True)
```

```python
sel_mod = SelectFromModel(Lasso(alpha=.005,random_state=0))
sel_mod.fit(X_train,y_train)
```

```python
sel_mod.get_support()
```

```python
selected_features = X_train.columns[(sel_mod.get_support())]

print('Total features: {}'.format((X_train.shape[1])))
print('Selected features: {}'.format(len(selected_features)))
print('Features which coefficient is shrunk to zero: {}'.format(np.sum(sel_mod.estimator_.coef_ == 0)))
```

```python
selected_features
```

```python
selected_features_df_train = X_train[selected_features]
selected_features_df_test = X_test[selected_features]
selected_features_df_train.head()
```

```python
selected_features_df_train.to_csv('selected_features_train.csv', index=False)
selected_features_df_test.to_csv('selected_features_test.csv', index=False)
```

```python
pd.Series(selected_features).to_csv('Selected_feature_list.csv',index=False)
```
